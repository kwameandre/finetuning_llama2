```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_excel('PIRS_LABELED_nonsubcategory.xlsx')

# Drop the specified columns

columns_to_drop = ['Event_ID', 'Near_Miss', 'Any_Harm']
df = df.drop(columns=columns_to_drop)

# Save the modified DataFrame to a new CSV file
print(df)
# Assuming df is your DataFrame
# Combine columns starting with 'Facilitators' into a single column 'Facilitators'
# Assume df is your DataFrame

# Select columns that start with 'Facilitators'
facilitators_columns = [col for col in df.columns if col.startswith('Facilitators')]

# Concatenate the selected columns along the row axis and join the values with a space
df['Facilitators'] = df[facilitators_columns].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)

# Drop the individual 'Facilitators' columns
df.drop(columns='Facilitators.1', inplace=True)
df.drop(columns='Facilitators.2', inplace=True)
# Now, your DataFrame has a single column named 'Facilitators' that combines the values from the original columns.


# Now, your DataFrame 'df' has the combined 'Facilitators' column.
#print(df)
#df.to_csv('facil2.csv', index=False)
```

```python
# Melt the DataFrame to transform it into long format
melted_df = pd.melt(df, id_vars=['Text'], value_vars=['Actions_Strategies','Risk_Challenge','Facilitators'], var_name='Label', value_name='Label_Value')

# Combine text with label title
melted_df['Text'] = melted_df['Text'] + ': ' + melted_df['Label']

# Drop the 'Label' column
final_df = melted_df.drop(columns=['Label'])

# Rename the columns
final_df.columns = ['text', 'label']

# Save the transformed DataFrame to a new CSV file
final_df.to_csv('test_all.csv', index=False)

```
